## 3. sqlalchemy함수에 있는 create_engine을 불러오겠다

```python
import psycopg2
from sqlalchemy import create_engine 
## sqlalchemy함수에서 create_engine를 불러오겠다(create_engine는 원래 있는것)
```

```python
mylocalpgUrl = "postgresql://postgres:postgres@192.168.110.13/postgres" ##  컴퓨터로 보내겠다
```

```python
engine_pg = create_engine(mylocalpgUrl) ## 연결해줄 방(링크)생성
```

```python
resultTableName = "pg_out333" ## 최종적으로 보낼 테이블 이름 다른이름으로 저장(pg_out333)
```

```python
data.to_sql(name=resultTableName,
           con=mylocalpgUrl, if_exists="replace", index=False) ## con : connection, 연결
```

```python
pd.read_sql(sql = "select * from pg_out", con = engine_pg) 

pd.read_sql_table(names="pg_out33", con=engine_pg)             ## 둘 다 같은표현!!!
```
