# 1. np.where방식

```python
import pandas as pd
```

```python
# CSV 파일을읽어Data Frame변수에저장하기
customerData= pd.read_csv("../dataset/customerData.csv")
```

```python
import numpy as np
```

```python
customerData.columns
```

```python
customerData["PRODUCTAGE_NEW"] = \
    np.where( customerData.PRODUCTAGE < 1, 1, 0 )
```

```python
## 이렇게도 가능하다~!
customerData["PRODUCTAGE_NEW"] = \
    np.where( customerData.PRODUCTAGE < 1, 1, \
            np.where(customerData.PRODUCTAGE < 2,2,3)) #1보다 작으면 1, 1보다 크면 다시 np~ 여기로 넘어감 그래서 2보다 작을땐 2, 그 외에는3
```

```python
customerData.loc[ (customerData.PRODUCTAGE < 1) &
                 (customerData.PRODUCTAGE_NEW != 1)]    #1보다 작은거 조회하니까 !=1 하는 반대되는 검증로직도 하는 습관 필요~!
```

# 2. 함수활용

```python
# 함수정의 1미만인 경우 : 1, 2미만 -> 2, 그 이외 3
```

```python
# 함수정의
#inputValue < 1 : 1
# 1 <= inputValue < 2 : 2
# inputValue >= 2 3
```

```python
def valueClustering(inValue):

    # inValue = 2

    outValue = inValue

    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    else: 
        outValue = 3
    return outValue
```

```python
customerData["PRODUCTAGE_NEW"] = customerData.PRODUCTAGE.apply(valueClustering)
```

```python
customerData
```
