## 합쳐보자

```python
import pandas as pd
import requests, bs4
from bs4 import BeautifulSoup # 태그 이쁘게 뜨기
import requests # 웹페이지 접속 요청
targetUrl = "https://www.daisomall.co.kr/deal/lowdeal_search.php?search_text=%EB%A7%88%EC%8A%A4%ED%81%AC&x=0&y=0"

resp = requests.get(url=targetUrl) ## response 200 = 정상

resp

resp.encoding = "utf-8" ## 인코딩
html = resp.text
htmlObj = BeautifulSoup(html, "html.parser") ## beautifulSoup 사용

htmlObj

maskList = []
priceList = []

## 전체페이지를 긁어오자

Totalpage = htmlObj.find(name = "div", attrs = {"class":"contents_center"})

Totalpage

mask=Totalpage.findAll(name = "a",attrs = {"class":"goods-title"})
for i in range(0, len(mask)):
## 이름을 긁어오자
    maskName=mask[i].attrs["title"]

    definedName=maskName.replace("<b>", "")

    defindeNameEnd=definedName.replace("</b>", "")

    ## 가격을 긁어오자

    price=Totalpage.findAll(name = "span",attrs = {"class":"low_goods_span2"})

    maskPrice=price[i]

    trueMaskPrice=maskPrice.text
    
    
    maskList.append(defindeNameEnd)
    priceList.append(trueMaskPrice)
    

pd.DataFrame(zip(maskList,priceList),columns=["이름","가격"])
```

```python
import pandas as pd
from selenium import webdriver 
from bs4 import BeautifulSoup # 태그 이쁘게 뜨기
import requests, bs4 # 웹페이지 접속 요청

options = webdriver.ChromeOptions()

#헤드리스 옵션 / 일반옵션    # 헤드리스란 ? 브라우저를 띄우지 않고 크롤링을 하는 것 
options.add_argument("window-size=1920*1080")

driverLoc = "../addon/chromedriver/chromedriver.exe"

driver = webdriver.Chrome(executable_path=driverLoc, options=options)

options.add_argument("window-size=1920*1080")

# 웹페이지 파싱 될 때까지 최대 3초 기다림
# driver.implicitly_wait(3)

targetUrl = "https://www.daisomall.co.kr/"

driver.get(targetUrl)

driver.page_source  # 페이지 소스를 보여주는 것

#타임딜 버튼을 xpath카피한걸 붙여넣고 변수를 지정하자
timeDealXpath = '//*[@id="header"]/div[4]/div/div[1]/ul/li[1]/a/img'

#클릭에 대한 행동정의
timeDeal = driver.find_element_by_xpath(timeDealXpath)

#타임딜 버튼 누르는 순간 다이소몰에서 타임딜 창 눌림
timeDeal.click()

# 브라우저 내 액션(텍스트 입력 및 검색)  -> 타임딜 검색창에 "마스크" 텍스트 입력 및 검색하기
from selenium.webdriver.common.keys import Keys
daisoSearchXpath = '/html/body/div[1]/div[2]/div[2]/div/div[1]/div/div[1]/form/div/input[1]' ## 타임딜의 검색창의 Xpath긁어온 것
daisoSearchBox = driver.find_element_by_xpath(daisoSearchXpath) # 검색에 대한 행동정의
searchKeyword = "마스크"
daisoSearchBox.send_keys(searchKeyword)
daisoSearchBox.send_keys(Keys.ENTER) # 실행하면 엔터버튼이 눌림

#현재 url확인용
driver.current_url

diso_source = driver.page_source   # 전체페이지 소스 가져오기 그걸 변수 diso_source에 담은 것
diso_source

## beautifulSoup 사용
htmlObj = BeautifulSoup(diso_source, "html.parser") 

# 전체 페이지를 긁어오자
totalPage = htmlObj.find(name="div", attrs={"class":"low_goods_list"})
Totalpage = htmlObj.find(name = "div", attrs = {"class":"contents_center"})

maskList = []
priceList = []

mask=Totalpage.findAll(name = "a",attrs = {"class":"goods-title"})
for i in range(0, len(mask)):
## 이름을 긁어오자
    maskName=mask[i].attrs["title"]

    definedName=maskName.replace("<b>", "")

    defindeNameEnd=definedName.replace("</b>", "")

    ## 가격을 긁어오자

    price=Totalpage.findAll(name = "span",attrs = {"class":"low_goods_span2"})

    maskPrice=price[i]

    trueMaskPrice=maskPrice.text
    
    
    maskList.append(defindeNameEnd)
    priceList.append(trueMaskPrice)
    

pd.DataFrame(zip(maskList,priceList),columns=["이름","가격"])

finalResult=pd.DataFrame(zip(maskList,priceList),columns=["이름","가격"])

#생성된 데이터프레임 csv파일로 저장
finalResult.to_csv("./middleTest.csv",encoding="ms949",index=False)

finalResult
```
