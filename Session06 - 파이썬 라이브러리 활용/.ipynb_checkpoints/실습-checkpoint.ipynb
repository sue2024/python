## 실습 :2015년도 자료만 활용하여
# x축: week, 이중축으로 qty, pro_percent를 차트로 그리세요

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
featuresData = pd.read_csv("../dataset/feature_regression_example.csv")
featuresData
```

```python
features2015 = featuresData.loc[ featuresData.YEAR == 2015 ]
features2015
```

```python
%matplotlib inline
xValue = features2015.WEEK
yValue = features2015.QTY
y2Value = features2015.PRO_PERCENT

fig = plt.figure(figsize=(10,5)) # 10은 x축, 길이는 5

ax1 = fig.add_subplot(1,1,1) # 2행1열로 영역을 나누고 그중에 첫번째영역에  ax1이라는 그림을 그리겠따
ax1.plot(xValue, yValue,"b--", label = "qty") # plot(선)을 그릴 때 라벨을 넣어줘야 한다~!
ax1.legend(loc=0)
plt.title('subplot') # 그래프의 이름
ax1.set_ylabel('qty') # 1번째 y축의 이름
ax1.set_xlabel('week')
ax2 = ax1.twinx() # x축을 같이 쓰겠다~!
ax2.set_ylabel('pro_percent') # 2번째 y축의 이름
ax2.plot(xValue, y2Value,"g--", label = "pro_percent" )
ax2.legend(loc=0)
```
